```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate the total number of schools
total_school = school_data['school_name'].count()

# Calculate the total number of students
total_student = student_data['Student ID'].count()

# Calculate the total budget
total_budget = school_data['budget'].sum()

# Calculate the average math score
average_math =  student_data['math_score'].mean()

# Calculate the average reading score
average_reading = student_data['reading_score'].mean()

# Calculate the percentage of students with a passing math score (70 or greater)
math_pass_percentage = 100*(student_data.loc[student_data['math_score'] >= 70]['Student ID'].count()/total_student)

# Calculate the percentage of students with a passing reading score (70 or greater)
reading_pass_percentage = 100*(student_data.loc[student_data['reading_score'] >= 70]['Student ID'].count()/total_student)

# Calculate the overall passing rate
overall_pass_percentage = (math_pass_percentage + reading_pass_percentage)/2

# Create a dataframe to hold the above results
district_summary = pd.DataFrame({"Total Schools" : [total_school], "Total Student" : [total_student], "Total Budget" : [total_budget],
                   "Average Math Score" :[average_math ], "Average Reading Score" :[average_reading ], 
                   "% Pass Math" : [math_pass_percentage], "% Pass Reading" : [reading_pass_percentage], 
                   "% Overall Passing Rate" :[overall_pass_percentage]})

# formatting district_summary data
district_summary["Total Student"] = district_summary["Total Student"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["Average Math Score"] = district_summary["Average Math Score"].map("{:.2f}".format)
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].map("{:.2f}".format)
district_summary["% Pass Math"] = district_summary["% Pass Math"].map("{:.2f}".format)
district_summary["% Pass Reading"] = district_summary["% Pass Reading"].map("{:.2f}".format)
district_summary["% Overall Passing Rate"] = district_summary["% Overall Passing Rate"].map("{:.2f}".format)

district_summary
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
# group by school name
group_school = school_data_complete.groupby(["school_name"])

# Data for School Type, Total Students, Total School Budget, Per Student Budget

per_student_budget = school_data["budget"]/school_data["size"]

school_type_pf = pd.DataFrame({"school_name" : school_data["school_name"], "School Type" : school_data["type"], 
                               "Total Students" : school_data["size"], "Total Budget" : school_data["budget"], 
                               "Per Student Budget" : per_student_budget } )

# Total Students for each School
average_math_school = group_school["math_score"].mean()

# Average Reading Score
average_reading_school = group_school["reading_score"].mean()

# % Passing Math
pass_math = school_data_complete.loc[school_data_complete["math_score"] >= 70]
pass_math_school = pass_math.groupby(["school_name"])["math_score"].count()
math_pass_rate = 100*(pass_math_school/total_student_school)

# % Passing Reading
pass_reading = school_data_complete.loc[school_data_complete["reading_score"] >= 70]
pass_reading_school = pass_reading.groupby(["school_name"])["reading_score"].count()
reading_pass_rate = 100*(pass_reading_school/total_student_school)

# Overall Passing Rate
overall_passing_rate = (math_pass_rate + reading_pass_rate)/2

# create a new data frame for school performing
school_performing_pf = pd.DataFrame({"Average Math Score" : average_math_school, "Average Reading Score" : average_reading_school, 
                                     "% Passing Math" : math_pass_rate, "% Passing Reading" : reading_pass_rate, 
                                     "% Overall Passing Rate" : overall_passing_rate})

school_performing_pf = pd.merge(school_type_pf, school_performing_pf, how="left", on=["school_name", "school_name"])


# sorting for best perforing school based on % Overall Passing Rate
best_school_performing = school_performing_pf.sort_values("% Overall Passing Rate", ascending=False)

# formatting for best perforing school based on % Overall Passing Rate
best_school_performing["Total Budget"] = best_school_performing["Total Budget"].map("${:,.2f}".format)
best_school_performing["Per Student Budget"] = best_school_performing["Per Student Budget"].map("${:.2f}".format)

best_school_performing.head(5)
```
